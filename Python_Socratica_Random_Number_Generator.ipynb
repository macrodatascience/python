The difference between **true random number generators(TRNGs)** and **pseudo-random number generators(PRNGs)** is that TRNGs use an unpredictable physical means to generate numbers (like atmospheric noise), and PRNGs use mathematical algorithms (completely computer-generated).

```python
import random
# The pseudo random generators of this module 
# It represents the uniform distribution i-e the prob of numbers chosen is equally spreadout

random.seed(9)

print(dir(random))
print(random.randint(1,10))

def my_random():
    # Random, Scale, Shift, Return
    # [0,1)
    return 4*random.random()+3

count=0
for i in range(100000):
    count+=my_random()
print(count/100000)

count=0
for i in range(100000):
    count+=random.uniform(3,7)
print(count/100000)
## random() function can be used to generate customised random number generators

## Generate random numbers between 3 and 7
## Step1 = Generate random numbers betwee 0 and 1
## Step2 = Multiply by 4
## Step3 = Shift the number by 3 (add 3)
```
