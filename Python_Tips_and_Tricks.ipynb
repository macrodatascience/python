```python
# Prime numbers

def is_prime(n):
  count=0
  for i in range(2, n//2+1):
    if n%i==0:
      return False
    count+=1
  return True
```

```python
is_prime(511121)
```

```python
def is_prime_one_liner(n):
  return 0 not in [n%i for i in range(2,n//2+1)]
```

```python
is_prime_one_liner(499)
```

```python
import time

start = time.time()
is_prime(511111)
end = time.time()
print(end-start)

start = time.time()
is_prime_one_liner(511111)
end = time.time()
print(end-start)
```

```python
# primes between two numbers

def is_prime_between(a,b):
  primes = []
  for i in range(a,b+1):
    if is_prime(i):
      primes.append(i)
  return primes
```

```python
is_prime_between(4,50)
```

```python
is_prime(47)
```

**# 1. List Comprehensions - Concise way to create lists in Python**

```python
nums = [2,4,6,8,19,12,312,123,2131,231,31,3,1,312,3,123,12,31,23,12,3,13,12,31,231,23,1]
evens = []
for num in nums:
  if num%2==0:
    evens.append(num)
print(evens)
```

```python
odds = [num for num in nums if num%2==1]
odds
```

```python
evens = [x for x in nums if x%2==0]
evens
```

```python
exponents = [x**2 for x in nums if x%2==0]
exponents
```

```python
# make all the first letters upper case

# using List Comprehension

words = ["new york", "moscow", "tokyo", "frankfurt", "mumbai", "bogota"]

[word.capitalize() for word in words]
```

```python
# using Map function

list(map(lambda word:word.capitalize(), words))
```

```python
# make all those letters starting with 'm' upper case

# using List Comprehension

words = ["new york", "moscow", "tokyo", "frankfurt", "mumbai", "bogota"]

[word.upper() if word.startswith('m') else word for word in words]
```

```python
[word.upper() for word in words if word.startswith('m')]
```

```python
nums = [2,4,5,6,12,12,12,312,32,32,3,213,21,312,31]

squared_nums = []

def square(x):
  return x**2

for i in nums:
  squared_nums.append(square(i))

print(squared_nums)  
```

```python
# list comprehension is more prevalent now than maps (as maps were useful when there was no list comprehension)

[i**2 for i in nums]
```

# Python Code Challenges

```python
def index_all(search_list,item):
  indices = []
  for i in range(len(search_list)):
    if search_list[i]==item:
      indices.append([i])
    elif isinstance(search_list[i],list):
      for index in index_all(search_list[i], item):
        indices.append([i]+index)
  return indices
```

```python
# Find all list items 
# from index_all import index_all
example = [[[1,2,3], 2, [1,3]], [1,2,3]]
index_all(example,2)
```

```python
index_all(example,3)
```

# waiting game

```python
# print a random amount of time 2-4 seconds 
# wait 4 seconds
# press enter to begin
# press enter after 4 seconds
# elapsed time: 4.213 seconds ?
```

```python
import random
print(random.randint(2,4))
```

```python
import time 
import random

def waiting_game():
  target = random.randint(2,4)
  print("\n Your target time is {} seconds...".format(target))
  input("---Press Enter to begin---")
  start = time.perf_counter()
  input("\n---Press enter again after {} seconds---".format(target))
  elapsed = time.perf_counter()-start 

  print("\nElapsed time: {0:.3f} seconds".format(elapsed))
  if elapsed == target:
    print("Unbelievable! Perfect timing")
  elif elapsed<target:
    print("({0:.3f} seconds too fast)".format(target - elapsed))
  else:
    print("({0:.3f} seconds too slow".format(elapsed-target))
```

```python
waiting_game()
```

**# Pickling is a process which serializes a Python object by converting into Byte stream that can be saved into a file. The inverse is unpickling it, that can be used to deserialize the byte stream back into the original object structure**

image.png

```python
import pickle

def save_dict(dict_to_save, file_path):
  with open(file_path, 'wb') as file:
    pickle.dump(dict_to_save, file)

def load_dict(file_path):
  with open(file_path, 'rb') as file:
    return pickle.load(pickle)
```

# Magic methods -- Dunder - Double underscore

```python
class Vector:
  def __init__(self, x, y):
    self.x = x
    self.y = y

v1 = Vector(5,10)
v2 = Vector(10,20)
v3 = v1 + v2
```

```python
class Vector:
  def __init__(self, x, y):
    self.x = x
    self.y = y
 
  def __add__ (self, other):
    return Vector(self.x + other.x, self.y + other.y)

  def __mul__(self, other):
    return Vector(self.x * other.x, self.y * other.y)

  def __rmul__(self, other):
    return Vector(self.y * other.y, self.x * other.x)

  def __repr__(self):
    return f"X: {self.x}, Y: {self.y}"

  def __call__(self):
    print(f"Hello, I am object {self}")

v1 = Vector(5,10)
v2 = Vector(10,20)
v3 = v1 + v2

v4 = v1 * v2  # v1__mul__v2 
v5 = v2 * v1

print(v3.x, v3.y)
print(v4.x, v4.y)

print(v3)  ## this uses the __repr__ method
print(v4)
print(v5)
# You have defined the dunder method so that the resulting product of two Data objects is a Data object itself

print(type(v4))

print(v2)
v2()
print(v2())  # if the __call__ method isn't defined then it will throw an error - TypeError: 'Vector' object is not callable
```
