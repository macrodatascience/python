```python
import logging

# Create and Configure a logger

# Setting up the logging systems using the basicConfig method
# Create the logger using the getLogger method

# How we configure the logger -- logger.level 

# force = True # Resets any previous configuration
# by default log file is in append mode
# if u want to overwrite, set the filemode = 'w'

LOG_FORMAT = "%(levelname)s %(asctime)s - %(message)s"
logging.basicConfig(filename = './lumberjack.log', 
                    level=logging.DEBUG, 
                    force=True, 
                    format = LOG_FORMAT,
                    filemode = 'w')
logger = logging.getLogger()      # root logger

# Test the logger
logger.debug("This is a debug message")
logger.info("Our first message")
logger.warning("This is a warning message")
logger.error("This is an error message")
logger.critical("This is a critical issue")
print(logger.level)
```

```python
import math
import logging
LOG_FORMAT = "%(levelname)s %(asctime)s - %(message)s"
logging.basicConfig(filename = './logfile.log',
                    level = logging.DEBUG,
                    format = LOG_FORMAT,
                    force = True,
                    filemode = 'w')
logger = logging.getLogger()

def quadratic_formula(a,b,c):
  """Return the solutions to the equation ax^2 + bx + c = 0"""
  logger.info("quadratic_formula({0}, {1}, {2}".format(a,b,c))
  # compute the discriminant
  logger.debug("# compute the discriminant")
  disc = b**2 - 4*a*c

  # compute the two roots
  logger.debug("# compute the two roots")
  root1 = (-b + math.sqrt(disc))/(2*a)
  root2 = (-b - math.sqrt(disc))/(2*a)

  # return the roots
  logger.debug("# return the roots")
  return (root1, root2)

roots1 = quadratic_formula(1,0,-4)
```
