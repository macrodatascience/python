```python
# Let's start by putting the following (completely fictional) share price lists into NumPy arrays

microsoft = [120.3, 121.7, 119.8, 122.5, 123.3]
apple = [124.1, 124.3, 124.2, 124.8, 124.3]
facebook = [110.2, 112.4, 116.8, 115.3, 119.6]
```

```python
# TASK 1: Import NumPy and create arrays

import numpy as np

microsoft = np.array(microsoft)
apple = np.array(apple)
facebook = np.array(facebook)
```

```python
type(microsoft), type(facebook), type(apple)
```

```python
# Now let's use some NumPy functions to learn some interesting statistical information

mean_msft = np.mean(microsoft)
mean_apple = np.mean(apple)
mean_facebook = np.mean(facebook)

# TASK 2A: Calculate the mean share price for each company, and store in suitably named variables.
#          Then print them out (in order - Microsoft, then Apple, then Facebook).


print("Mean values: Microsoft -> {:.2f}, Apple -> {:.2f}, Facebook -> {:.2f}".format(mean_msft, mean_apple, mean_facebook))

# TASK 2B: Calculate the standard deviation of share prices for each company using a NumPy function,
#          and store in suitably named variables. Then print them out (in order - Microsoft, then Apple, then Facebook).

std_microsoft = np.std(microsoft)
std_apple = np.std(apple)
std_facebook = np.std(facebook)

print("Std. deviation values: Microsoft -> {:.2f}, Apple -> {:.2f}, Facebook -> {:.2f}".format(std_microsoft, std_apple, std_facebook))

# TASK 2C: Calculate the percentage change in share price from the first price to the last price, and store 
#          in suitably named variables. Then print them out (in order - Microsoft, then Apple, then Facebook).

def daily_price_chg(company):
  result = (np.diff(company)/company[1:])*100
  return result

def price_chg(company):
  result = (company[len(company)-1]-company[0])/company[0]
  return result

microsoft_price_chg = (microsoft[len(microsoft)-1]-microsoft[0])/microsoft[0]
apple_price_chg = (apple[len(apple)-1]-apple[0])/apple[0]
facebook_price_chg = (facebook[len(facebook)-1]-facebook[0])/facebook[0]

print("% change values: Microsoft -> {:.4f}%, Apple -> {:.4f}%, Facebook -> {:.4f}%".format(microsoft_price_chg*100, apple_price_chg*100, facebook_price_chg*100))

microsoft_price_chg = price_chg(microsoft)
apple_price_chg = price_chg(apple)
facebook_price_chg = price_chg(facebook)

print("% change values: Microsoft -> {:.4f}%, Apple -> {:.4f}%, Facebook -> {:.4f}%".format(microsoft_price_chg*100, apple_price_chg*100, facebook_price_chg*100))

```

```python
# Now we're going to load our NumPy arrays into a Pandas Dataframe
# The 5 values in each company's array is the share price for the first 5 months of 2018

# TASK 3: Import Pandas and load the arrays into a dataframe with appropriate row and column labels, and then
#         output the dataframe by simply typing its variable name on a new line

import pandas as pd

df = pd.DataFrame([microsoft, apple, facebook], index=['Microsoft', 'Apple', 'Facebook'], columns = ['Jan 2018', 'Feb 2018', 'Mar 2018', 'Apr 2018', 'May 2018'])
```

```python
# Finally, let's draw some interesting graphs! 
# You will need to use some of the statistical values you calculated earlier in order to do this

# TO DO: Import Matplotlib's Pyplot submodule

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# TASK 4A: Draw a line graph for one of the companies, showing their share prices over time, using the dataframe.
#          Provide suitable labels. Hint: using your_dataframe_variable.loc["your_row_label", :] will extract a single row.

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(df.loc["Microsoft",:])

plt.title("Microsoft Share Price", fontdict = font1)
plt.xlabel("Months in 2018", fontdict = font2)
plt.ylabel("Share Price", fontdict = font2)
plt.show()

plt.plot(df.loc["Apple",:])

plt.title("Apple Share Price", fontdict = font1)
plt.xlabel("Months in 2018", fontdict = font2)
plt.ylabel("Share Price", fontdict = font2)
plt.show()

plt.plot(df.loc["Facebook",:])

plt.title("Facebook Share Price", fontdict = font1)
plt.xlabel("Months in 2018", fontdict = font2)
plt.ylabel("Share Price", fontdict = font2)
plt.show()
```

```python
# TASK 4B: Draw a bar chart, using any data structure, comparing each company's value for any statistic of your choice 
#          from Task 2 (A, B or C)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
shares = ['Microsoft', 'Apple', 'Facebook']
std_shares = [std_microsoft, std_apple, std_facebook]
ax.bar(shares,std_shares)
plt.title("Comparison of Standard Deviation for Microsoft, Apple and Facebook", fontdict = font1)
plt.ylabel("Standard Deviation", fontdict = font2)
plt.show()
```

```python
# TASK 4B: Draw a bar chart, using any data structure, comparing each company's value for any statistic of your choice 
#          from Task 2 (A, B or C)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
shares = ['Microsoft', 'Apple', 'Facebook']
percent_return = [microsoft_price_chg*100, apple_price_chg*100, facebook_price_chg*100]
ax.bar(shares,percent_return)
plt.title("Comparison of % Returns for Microsoft, Apple and Facebook", fontdict = font1)
plt.ylabel("Percentage Return", fontdict = font2)
plt.show()
```
